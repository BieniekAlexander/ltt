```python
# imports
import pandas as pd
import numpy as np
from scraping.canto_dict.canto_dict_client import get_fact_from_characters
import json
import logging

logging.basicConfig(level="INFO")
```

```python
hsk1_df = (
    pd
    .read_csv('data/hsk1_simplified.csv', comment='#')
    .fillna(np.nan).replace([np.nan], [None])
)
```

```python
hsk1_fact_dict = {}
character_set_names = ['traditional','simplified']

for index in hsk1_df.index:
    # scrape a radical from cantodict
    characters = hsk1_df['characters'][index]
    english = hsk1_df['meaning'][index]
    logging.info(characters)
    fact = get_fact_from_characters(characters)
    
    if fact is None:
        logging.warning(f"fact not scraped for {characters}")
    else:
        hsk1_fact_dict[characters] = fact
```

```python
# write to file
with open('data/hsk1.json', 'w') as file:
     file.write(json.dumps(hsk1_fact_dict))
```
